```python
%%mml
CREATE POPULATION FOR "mammals_full" WITH SCHEMA (
    -- Use the guesses from the previous cell for all variables.
    GUESS STATTYPES OF (*);

    -- Manually override incorrectly guessed statistical types.
    SET STATTYPES OF
        "Body Temperature",
        "Body length, nose to tail",
        "Male Tenure",
        "Reproductive Skew",
        "Testes Mass",
        "age at eye opening",
        "age at first birth",
        "dispersal age",
        "growth rate",
        "initial mortality rate",
        "head-body length",
        "hind tarsus length",
        "human population density",
        "life span",
        "mortality rate doubling time",
        "outer ear length",
        "prenatal development duration",
        "water salinity"
    TO NUMERICAL;
);
```

```python
%%bql
.interactive_pairplot --population=mammals_full 
SELECT
    "age at first birth",
        "head-body length",
        "hind tarsus length"
FROM mammals_full
```

```python
%mml CREATE GENERATOR FOR "mammals_full" USING loom;
```

```python
%mml INITIALIZE 10 MODELS IF NOT EXISTS FOR "mammals_full";
```

```python
%%bql
CREATE TABLE dependencies AS
ESTIMATE
    DEPENDENCE PROBABILITY AS "depprob"
FROM PAIRWISE VARIABLES OF mammals_full;
```

```python
%bql .interactive_heatmap SELECT name0, name1, depprob FROM dependencies;
```
